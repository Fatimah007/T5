## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
numerical_columns = df[['TrafficIndexLive','JamsDelay']]
correlation_matrix = numerical_columns.corr()
print(correlation_matrix)
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Datetime', y='JamsDelay')
plt.title('bar chart')
plt.xlabel('Datetime')
plt.ylabel('JamsDelay')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Datetime', y='JamsCount')
plt.title('histogram chart')
plt.xlabel('Datetime')
plt.ylabel('JamsCount')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Datetime', y='JamsCount')
plt.title('scatter plots')
plt.xlabel('Datetime')
plt.ylabel('Jamscount')
plt.show()
```
